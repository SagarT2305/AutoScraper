```python
from autoscraper import AutoScraper

url = 'https://www.flipkart.com/search?q=mobiles'
```

```python
wanted_list = ['APPLE iPhone SE (Red, 128 GB)','₹29,999', '4.5']

scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar(url, grouped = True)
```

```python
scraper.set_rule_aliases({'rule_gvcm':'Product_Name',
                          'rule_u9xt':'Price',
                          "rule_2g0c":'Stars' })
scraper.keep_rules(['rule_gvcm','rule_u9xt','rule_2g0c'])
scraper.save('Mobiles')
```

```python
scraper.load('Mobiles')
```

```python
laptops = scraper.get_result_similar('https://www.flipkart.com/search?q=laptops',group_by_alias=True )
```
